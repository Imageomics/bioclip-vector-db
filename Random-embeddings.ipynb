```python
from pymilvus import connections, FieldSchema, CollectionSchema, DataType, Collection, Index
```

```python
# Connect to Milvus
connections.connect("default", host="localhost", port="19530")
```

```python
# Define the collection name and parameters
collection_name = "random_embeddings"
dimension = 512
num_embeddings = 10_000_000
batch_size = 10_000  # Process in batches of 10,000 embeddings
```

```python
# Define the schema
fields = [
    FieldSchema(name="embedding", dtype=DataType.FLOAT_VECTOR, dim=dimension),
    FieldSchema(name="id", dtype=DataType.INT64, is_primary=True, auto_id=True)
]
schema = CollectionSchema(fields, description="Collection of random embeddings")
```

```python
# Create the collection
collection = Collection(name=collection_name, schema=schema)
```

```python
# Create an index with Inner Product (IP) distance metric
index_params = {
    "index_type": "IVF_FLAT",
    "metric_type": "IP",
    "params": {"nlist": 128}
}
collection.create_index(field_name="embedding", index_params=index_params)
```

```python
# Generate and insert random embeddings in batches
num_batches = num_embeddings // batch_size
```

```python
for i in range(num_batches):
    embeddings = np.random.random((batch_size, dimension)).astype(np.float32)
    collection.insert([embeddings])
    print(f"Inserted batch {i+1}/{num_batches}")
```
